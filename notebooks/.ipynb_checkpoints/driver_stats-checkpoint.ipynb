```python
import requests
import pandas as pd
import json
import matplotlib.pyplot as plt # Needed to display the plots
from datetime import datetime, timedelta
import ipywidgets as widgets
from IPython.display import display
```

```python
class OpenF1Request:
    
    def __init__(self, endpoint, api_version = 'v1', **kwargs):
        self.base_url = 'https://api.openf1.org/'
        
        self.api_version = api_version
        
        self.allowed_endpoints = {'car_data':'Some data about each car, at a sample rate of about 3.7 Hz.', 
                         'championship_drivers':'Provides championship standings for drivers. Only available for race sessions.', 
                         'championship_teams':'Provides championship standings for teams. Only available for race sessions.',
                         'drivers':'Retrieve detailed information about the drivers participating in a specific session.',
                         'intervals':'Fetches real-time interval data between drivers and their gap to the race leader. Available during races only, with updates approximately every 4 seconds.',
                         'laps':'Provides detailed information about individual laps.',
                         'location':'The approximate location of the cars on the circuit, at a sample rate of about 3.7 Hz. Useful for gauging their progress along the track, but lacks details about lateral placement — i.e. whether the car is on the left or right side of the track. The origin point (0, 0, 0) appears to be arbitrary and not tied to any specific location on the track.',
                         'meetings':'Provides information about meetings. A meeting refers to a Grand Prix or testing weekend and usually includes multiple sessions (practice, qualifying, race, ...). Meetings are updated every day at midnight UTC.',
                         'overtakes':'Provides information about overtakes. An overtake refers to one driver (the overtaking driver) exchanging positions with another driver (the overtaken driver). This includes both on-track passes and position changes resulting from pit stops or post-race penalties. This data is only available during races and may be incomplete.',
                         'pit':'Provides information about cars going through the pit lane.',
                         'position':'Provides driver positions throughout a session, including initial placement and subsequent changes.',
                         'race_control':'Provides information about race control (session status, racing incidents, flags, safety car, ...).',
                         'sessions':'Provides information about sessions. A session refers to a distinct period of track activity during a Grand Prix or testing weekend (practice, qualifying, sprint, race, ...). Sessions are updated every day at midnight UTC.',
                         'session_result':'Provides standings after a session. This data becomes available a few minutes after the official results are published on the official Formula 1 website.',
                         'starting_grid':'Provides the starting grid for the upcoming race. This data becomes available a few minutes after the official results are published on the official Formula 1 website.',
                         'stints':'Provides information about individual stints. A stint refers to a period of continuous driving by a driver during a session.',
                         'team_radio':'Provides a collection of radio exchanges between Formula 1 drivers and their respective teams during sessions. Please note that only a limited selection of communications are included, not the complete record of radio interactions.',
                         'weather':'The weather over the track, updated every minute.'}
        
        self.open_f1_api_url = self.base_url + self.api_version
        
        self.params = {}
        for key, value in kwargs.items():
            self.params[key] = value
    
        if endpoint in self.allowed_endpoints:
            self.endpoint = endpoint
            self.request_url = self.open_f1_api_url + '/' + self.endpoint
            print('SUCCESS: Endpoint found')
            print(f'Name: {endpoint}')
            print(f'URL: {self.request_url}')
            print(f'Description: \n{self.allowed_endpoints[endpoint]}')
        else:
            print(f'The provided endpoint of "{endpoint}" is not allowed')
            print(f'Please select an endpoint from the following allowed endpoints: \n {self.allowed_endpoints}')
            return
    
    def send_api_request(self):
        print(f'Sending request to OpenF1...')
        
        response = requests.request("GET", url=self.request_url, params=self.params)
        self.response = response
        self.response_json = self.response.json()
        self.response_dataframe = pd.DataFrame(response.json())

        print(f'OpenF1 Response Code: {response.status_code}')
```

```python
class F1Sessions:

    def __init__(self):
        self.sessions_request = OpenF1Request('sessions')
        self.sessions_request.send_api_request()
        available_races = {}
        available_quali = {}
        available_practice = {}
        for ii, item in enumerate(self.sessions_request.response_json):
            if item['session_type'].lower() in ('race'):
                race_name_year = item['circuit_short_name'] + ' ' + item['session_name'] + ' ' + str(item['year'])
                race_session_key = item['session_key']
                race_info = {race_name_year:race_session_key}
                available_races.update(race_info)
            # elif item['session_type'].lower() in ('qualifying'):
            #     quali_name_year = item['circuit_short_name'] + ' ' +  + str(item['year'])
            #     race_session_key = item['session_key']
            #     race_info = {race_name_year:race_session_key}
            #     available_races.update(race_info)
        self.available_races = available_races            

        
```

```python
f1_sessions = F1Sessions()
race_dropdown = widgets.Dropdown(
    options=f1_sessions.available_races,
    description='Select a Race:',
)

display(race_dropdown)
```

```python
#Driver Data
driver_data = OpenF1Request('drivers', session_key = str(race_dropdown.value))
driver_data.send_api_request()
```

```python
#Laps Data
laps_data = OpenF1Request('laps', session_key = str(race_dropdown.value))
laps_data.send_api_request()
```

```python
laps_df = laps_data.response_dataframe
```

```python
fast_laps_df = laps_df[['driver_number','lap_duration']].groupby('driver_number').min()
```

```python
driver_df = driver_data.response_dataframe[['full_name', 'driver_number', 'team_name']]
```

```python
driver_fast_lap_df = pd.merge(left=driver_df, right=fast_laps_df, how='left', on='driver_number')
```
